Since there are 284315 0 values and 492 1 values , in order to avoid over fitting of data beacause of very number of 0 values compared to 1 values2

```python
legitimate_sample= legitimate.sample(n=492)
legitimate_sample
```

```python
new_df=pd.concat([legitimate_sample,fraud],axis=0)
```

```python
new_df
```

```python
new_df['Class'].value_counts()
```

Modeling


```python
x= new_df.drop(['Class'],axis=1)
y= new_df['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred= model.predict(x_test)
```

```python
accuracy= accuracy_score(y_test,y_pred)
```

```python
print(accuracy)
```

```python
from sklearn.metrics import  classification_report
print(classification_report(y_test,y_pred))
```
